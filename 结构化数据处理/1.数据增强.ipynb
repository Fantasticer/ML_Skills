### 二.删除操作
:param df: 输入的pandas  
:param cols: 需要被操作的cols，默认None，对所有列操作  
:param percent: 被操作的数量占比，按列（format_type=1）操作时生效，默认0.1  
:param nums: 被操作的数量，按行（format_type=0）操作时生效，默认2  
:param format_type: 0表示按行操作（速度慢），1表示按列操作（速度快），默认1

#### 按行置空

```python
dag.delete(df,nums=3,format_type=0).head(10)
```

### 三.替换操作
:param df:输入数据  
:param cols:被操作的列名称,默认None,对所有列操作  
:param percent: 被操作的数量占比，按列（format_type=1）操作时生效,默认0.1  
:param nums: 被操作的数量，按行（format_type=0）操作时生效,默认2  
:param format_type: 0表示按行操作（速度慢），1表示按列操作（速度快）,默认1  
:param replace_type:random表示随机替换，normal表示先建立高斯分布然后再抽样,默认random  
:param null_first:是否优先替换空值（相当于空值填充）,默认False  
:param class_label:是否按照class_label来操作,默认None   

#### 按列随机替换

```python
dag.replace(df).head(10)
```

#### 按行随机替换

```python
dag.replace(df,format_type=0).head(10)
```

#### 优先替换空值

```python
dag.replace(df,null_first=True).head(10)
```

#### 修改替换算法：高斯采样
先用当前特征训练一个高斯分布，然后再从该高斯分布随机采样做替换   
**注意，必须要是数值型**

```python
dag.replace(df,null_first=True,replace_type="normal",cols=['Age','Fare']).head(10)
```

#### 限制替换范围：组内替换
通过`class_label`为每一条数据赋予一个组id，所有的替换操作都被限制在同一组内，比如回归任务中，将数据分为高分、中等分、低分组，组内的数据彼此替换

```python
#将数据随机定义为两组
class_label=[0 if random.random()<0.5 else 1 for _ in range(0,df.shape[0])]
```

```python
dag.replace(df,class_label=class_label).head(10)
```
